```python
import numpy as np
import pandas as pd 
import time
import tracemalloc
import math
import matplotlib.pyplot as plt

####################################################################
# Notation of parameters (nm, ps, kg/mol)
# SDE:  x''+f*x'+k(x-v0*t)=U*sin(x/L)+ F(mu,sigma)

mu=0                    # fluctuation by average, always zero
sigma=0.020  # thermal fluctuation of force:  10^3kJ/(nm*mol)
dt = 0.1                # ps, step length
k=1               #  fitting parameter , 10^3kJ/(nm*nm*mol)
U=0.3558                # MJ/mol, potential energy
L=0.8                    #nm, induced by apl=0.62 nm*nm, L=0.8 nm
f=1                        # 1/ps, viscocity, fitting parameter
N=20000000               # number of steps
v0=0.001            # nm/ps, pulling velocity
####################################################################
AVR=[]
for i in range(1,10):
    # N=N*20           # to converge
    x = [0] * 2         #for efficiency
    v = [0] * 2  
    a = [0] * 2 
    t = [0] * 2
    #v0 = v0/10
    avr = 0
    rng = np.random.default_rng()
    rng
    rng.normal()
    dW =rng.normal();      # first approximation outside the loop ...

    for j in range(1, N):
        rng = np.random.default_rng()
        dW = sigma*math.sqrt(dt)*rng.normal();  # general increment
        out = k*(v0*j*dt-x[0])
        a[1] = out+U*np.sin(x[0]/L)+dW-f*v[0]
        v[1] = v[0] + a[1]*dt
        x[1] = x[0] + v[1]*dt + 0.5*a[1]*dt*dt  # 2nd order Verlet integration
        x[0]=x[1]
        v[0]=v[1]
        avr=avr+out/N
        if(j%(N//10)==0):
            print(x[1],v[1])
    print('N',N)
    print('v0',v0,'AVRf=',avr)
    AVR.append(avr)
    del x,v,a,t,out,avr,rng

X=0
for i in range(0,len(AVR)):
    X=X+AVR[i]/len(AVR)
print(X, '##',AVR)
```
